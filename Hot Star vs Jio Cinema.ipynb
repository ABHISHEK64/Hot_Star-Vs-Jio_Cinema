## Upgrade Patterns
#### What are the most common upgrade transitions (e.g., Free to Basic, Free to VIP, Free to Premium) for LioCinema and Jotstar? How do these differ across platforms?

```python
user_base['upgrade']=user_base.apply(lambda x:x['new_subscription_plan']!=x['subscription_plan'] and pd.notna(x['plan_change_date']),axis=1)
user_base['upgrade_transition']=user_base['subscription_plan']+'->'+user_base['new_subscription_plan']
user_base.head()
```

```python
upgrades = user_base[user_base['upgrade']]
upgrade_counts = upgrades.groupby(['Streaming_Patform', 'upgrade_transition'])['user_id'].count().reset_index()
upgrade_counts = upgrade_counts.sort_values(by=['Streaming_Patform', 'user_id'], ascending=[True, False])
upgrade_counts
```

```python
plt.figure(figsize=(10, 6))
ax=sns.barplot(data=upgrade_counts, y='upgrade_transition', x='user_id', hue='Streaming_Patform', palette='coolwarm')
for container in ax.containers:
    ax.bar_label(container,fmt='%d',label_type='edge',fontsize=8, padding=3)
plt.xlabel("Number of Upgrades")
plt.ylabel("Upgrade Transition")
plt.title("Most Common Upgrade Transitions: JioCinema vs. Hotstar")
plt.legend(title="Streaming Platform")
plt.show()
```

##  Revenue Analysis


```python
user_base.head()
```

```python
subscription_prices = {
    'Jio_Cinema': {'Basic': 199, 'Premium': 399, 'Free': 0},
    'Hot_star': {'VIP': 299, 'Premium': 499, 'Free': 0}
}
user_base['Monthly_price']=user_base.apply(lambda x:subscription_prices[x['Streaming_Patform']].get(x['subscription_plan'],0),axis=1)
user_base.head()
```

```python
user_base['last_active_date'] = pd.to_datetime(user_base['last_active_date'],format="mixed")
user_base['plan_change_date'] = pd.to_datetime(user_base['plan_change_date'],format="mixed")
user_base['Plan_end_date']=user_base[['plan_change_date', 'last_active_date']].min(axis=1)
```

```python
user_base['Plan_end_date']=user_base['Plan_end_date'].fillna(pd.to_datetime("2024-12-31"))
```

```python
user_base['months_active'] = ((user_base['Plan_end_date'] - user_base['subscription_date']) / np.timedelta64(1, 'M')).round()
user_base.head()
```

```python
user_base['Total_Revenue']=user_base['months_active']*user_base['Monthly_price']
revenue_Analysis=user_base.groupby(['Status_User','subscription_plan','Streaming_Patform','subscription_date_month'])['Total_Revenue'].sum().reset_index()
revenue_Analysis
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(data=revenue_Analysis, x='subscription_date_month', y='Total_Revenue',hue='Streaming_Patform')
plt.xlabel("Date_Time")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)
plt.title("Total Revenue month for Two Platforms in 2024")
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=revenue_Analysis, x='subscription_plan', y='Total_Revenue',hue='Streaming_Patform')
plt.xlabel("subscription_plan")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)
plt.title("Total Revenue of diffrent subscription_plan for Two Platforms in 2024")
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=revenue_Analysis, x='Status_User', y='Total_Revenue',hue='subscription_plan')
plt.xlabel("subscription_plan")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)
plt.title("Total Revenue of diffrent subscription_plan for Two Platforms in 2024")
plt.grid(True)
plt.legend()
plt.show()
```
