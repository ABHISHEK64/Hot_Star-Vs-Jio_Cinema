```python
# Load and process the file to extract data while removing the INSERT statements
file_path = r'C:\Users\abhis\OneDrive\Documents\INSERT_INTO `content_consumption`_jio.txt'

# Read the file content
with open(file_path, "r", encoding="utf-8") as file:
    lines = file.readlines()

# Extract only the VALUES portion from the INSERT statement
data_lines = []
for line in lines:
    if line.strip().upper().startswith("INSERT INTO"):
        # Extract the part after VALUES keyword
        values_part = line.split("VALUES", 1)[-1].strip(" ;\n")
        data_lines.append(values_part)

# Join and clean up the extracted values
clean_data = "".join(data_lines).strip("()")

# Replace "),(" with newline to structure as CSV
csv_data = clean_data.replace("),(", "\n")

# Add column headers (assuming column names from MySQL structure)
csv_output = "user_id,device_type,total_watch_time_mins\n" + csv_data

csv_file_path = r'C:\Users\abhis\OneDrive\Documents\data_cleaning_steps\content_consumption.csv'
with open(csv_file_path, "w", encoding="utf-8") as csv_file:
    csv_file.write(csv_output)


```
